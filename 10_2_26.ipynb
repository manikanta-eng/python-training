````python
def absolute(fun):
  def updater(v1,v2):
    if v1>v2:
      return fun(v1,v2)
    else:
      return fun(v2,v1)
  return updater

def diff(num1,num2):
  return num1-num2
d1=diff(10,5)
print(f'{d1=}')
abso=absolute(diff)
d2=abso(10,15)
print(f'{d2=}')

````
